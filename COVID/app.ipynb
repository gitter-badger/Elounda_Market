```python
from COVID import plot
import pandas as pd
import requests

# MAKE DF Reports Viewable
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# -------------------- DATAFRAME FUNCTION --------------------
def prepare_dataframe():
    url = 'https://covid.ourworldindata.org/data/ecdc/full_data.csv'
    r = requests.get(url)
    with open('covid_data.csv', 'wb') as f:
        f.write(r.content)
    df =pd.read_csv('covid_data.csv')
    return df
```

```python
# -------------------- DATAFRAME --------------------
covid_data = prepare_dataframe()
```

```python
# -------------------- GET ALL COUNTRIES --------------------
countries = covid_data.location.unique()
```

```python
# -------------------- SELECT WORLD --------------------
covid_data_world = covid_data[covid_data.location == 'World']
covid_data_world.new_deaths.sum()

# -------------------- SELECT GREECE --------------------
covid_data_greece = covid_data[covid_data['location'] == 'Greece']

# -------------------- PLOT GREECE --------------------
plot.plot_greece_graph(covid_data_greece)
```
